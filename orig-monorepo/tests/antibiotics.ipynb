```python
import numpy as np

from api.antibiotics import (
    simulate_antibiotic_response,
    get_MIC_curve,
    get_PAP_curve,
    get_single_cell_trajectories,
    list_available_parameters,
    run_parameter_scan
)
from api.plots import (
    plot_mic_curve,
    plot_pap_curve,
    plot_antibiotic_response,
    plot_single_cell_trajectories,
    plot_parameter_scan
)
```

```python
# 1. Simulate mecillinam response
response = simulate_antibiotic_response({"dose": 2.0})
plot_antibiotic_response(response)
```

```python
# 2. MIC curve simulation
mic_curve = get_MIC_curve({"sensitivity": 1.5})
plot_mic_curve(mic_curve)
```

```python
# 3. PAP curve simulation
pap_curve = get_PAP_curve({"mic": 4.0})
plot_pap_curve(pap_curve)
```

```python
# 4. Single-cell trajectories
trajectories = get_single_cell_trajectories(5, {"rate": 0.25})
plot_single_cell_trajectories(trajectories)
```

```python
# 5. List available parameters
available_params = list_available_parameters()
print(available_params)
```
