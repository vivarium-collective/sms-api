```python
from typing import Iterable
from collections import defaultdict
import numpy as np

def distribute_frames(start_time: float, duration: float, framesize: float, t: Iterable[float]):
    frames = np.arange(start_time, duration, framesize)
    keys = list(map(str, t))
    chunks = np.array_split(frames, len(keys))  # Evenly splits into len(t) chunks
    r = {k: list(v) for k, v in zip(keys, chunks)}
    return r

def test_frames():
    start = 0.1111
    end = 11.11
    fs = 3
    step = 0.000001
    t = np.arange(start, end, step)
    return distribute_frames(start, end, fs, t)

```

```python
def get_save_times(start, end, framesize, t):
    frames = np.arange(start, end, framesize)
    hist = np.histogram(frames, bins=t, range=None, density=None, weights=None)
    h = list(zip(hist[1], hist[0]))
    save_times = []
    for i, v in h:
        if v > 0:
            save_times.append(i.tolist())
    return save_times


def test_get_save_times():
    start = 0.1111
    end = 11.11
    fs = 0.1
    step = 0.000001
    t = np.arange(start, end, step)
    return get_save_times(start, end, fs, t)


test_get_save_times()
```
